```python
def getUrl(start,keyword='olympic'):
    url = f"https://www.washingtonpost.com/newssearch/?query={keyword}&sort=Date&datefilter=60%20Days&contenttype=Article&startat="+ str(start) + "&spellcheck"
    return url
```

```python
import pandas as pd
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

if not os.path.exists('./outputs'):
    os.mkdir('./outputs')
driver_path  = './chromedriver/chromedriver.exe'
comment_path = 'https://www.washingtonpost.com/comments?storyUrl='
driver = webdriver.Chrome(driver_path)
columns = ['title','link','content','date','reviews']
df = pd.DataFrame([],columns = columns) # review [ 'sadfasdf', 'asfasdf',...]
n_arts = 825
rows = 20 


# lists = driver.find_elements_by_class_name('pb-search-wrapper').find_elements_by_class_name('pb-search-wrapper')
import math
niters = math.ceil(n_arts/20)
for x in range(20):

    driver.get(getUrl(x*rows))
    newslist = driver.find_elements_by_css_selector('a[data-ng-bind-html="doc.headline"]')
    print('newslist nums',len(newslist))
    titles = []
    links  = []
    
    for page in newslist:
        titles.append(page.text)
        links.append(page.get_attribute('href'))
        
        
    for i,page in enumerate(newslist):
        
        print(titles[i])
        try:
            news_link = links[i]
            comment = comment_path + news_link
            driver.get(comment)
            # iframe searching problem
            src    = driver.find_elements_by_tag_name('iframe')
            driver.get(src[0].get_attribute('src'))
        except:
            print('continue')
            continue
            
        time.sleep(1)
        # number of comments 
        num_reviews = driver.find_elements_by_css_selector('span[class="Counter-text-58caffde738d519a54562205d642c007"]')[0].text
        num_reviews = int(num_reviews)
        print(num_reviews)
        
        comments = []
        if num_reviews != 0 :
            while True:
                buttons = driver.find_elements_by_css_selector('button[id="comments-loadMore"]')
                if len(buttons) != 0:
                    buttons[0].click()
                    time.sleep(1) # wait to load comments
                else:
                    break
                
            comment_tag_path = 'div[class="HTMLContent-root-5770ce4668399900d87c06ad10ba71a5 htmlContent-root-fe3506b39eaecd1a685ceda5e1292c4d coral coral-content coral-comment-content"]'
            comment_tags = driver.find_elements_by_css_selector(comment_tag_path)
            print('comments_tag num: ',len(comment_tags))
            for com in comment_tags:
                
                comments.append(com.text)
        else:
            continue
        
        row = [titles[i],None,None,None,comments] # ['title','link','content','date','reviews']
        df = df.append(dict(zip(columns,row)),ignore_index=True)
        #     print(newslist[0].text) # drivr 가 focused to another page
    df.to_csv(f'./outputs/washington{x}.csv')
    df = pd.DataFrame([],columns = columns)
    print('one page list is ended',"="*50)
print("the end")
driver.close()
```
